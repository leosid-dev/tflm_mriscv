```python
import numpy as np
```

```python
#### RELU #### 
np.random.seed(0)

def gen_rand_signed(bits=8, size=256):
    return np.random.randint(low=-2**(bits-1), high=2**(bits-1)-1, size=size)

input_data = gen_rand_signed(bits=8, size=2**8)
output_ref_data = np.maximum(input_data, 0)
print("Input:", repr(input_data))
print("Relu:", repr(output_ref_data))
output_ref_data = np.minimum(np.maximum(input_data, 0), 6)
print("Relu6:", repr(output_ref_data))
```

```python
#### ADD ####

def get_quantization_params(mini, maxi, scale=None):

    qmin = -128
    qmax = 127
    
    scale = (maxi-mini)/(qmax-qmin)
    zero_point = qmin-mini/scale
    
    if (zero_point < qmin):
        zero_point = qmin
    elif (zero_point > qmax):
        zero_point = qmax
    else:
        zero_point = round(zero_point)
    
    print(mini, maxi, scale, zero_point)

np.random.seed(0)
input_data1 = np.round(np.random.uniform(-4, 6, 128), 2)
input_data2 = np.round(np.random.uniform(-3, 2, 128), 2)
output_data = input_data1 + input_data2
print("Input1:", repr(input_data1))
print("Input2:", repr(input_data2))
print("Output:", repr(output_data))

get_quantization_params(-4,6)
get_quantization_params(-3,2)
get_quantization_params(-7,8)
```

```python
np.random.seed(0)
input_data1 = np.round(np.random.uniform(-2, 1, 128), 2)
input_data2 = np.round(np.random.uniform(-1, 4, 128), 2)
output_data = np.minimum(np.maximum(input_data1 + input_data2, -1), 1)
print("Input1:", repr(input_data1))
print("Input2:", repr(input_data2))
print("Output:", repr(output_data))

get_quantization_params(-2.5,1.5)
get_quantization_params(-1.5,4.5)
get_quantization_params(-1.5,1.5)
```

```python
#### MUL ####

def get_quantization_params_from_data(data, tolerance=1.5):
    mini = min(min(data), 0) * tolerance
    maxi = max(max(data), 0) * tolerance
    
    qmin = -128
    qmax = 127
    
    scale = (maxi-mini)/(qmax-qmin)
    zero_point = qmin-mini/scale
    
    if (zero_point < qmin):
        zero_point = qmin
    elif (zero_point > qmax):
        zero_point = qmax
    else:
        zero_point = round(zero_point)
    
    print(mini, maxi, scale, zero_point)

np.random.seed(1)
input_data1 = np.round(np.random.uniform(-2, 1, 128), 2)
input_data2 = np.round(np.random.uniform(-1, 4, 128), 2)
output_data = input_data1 * input_data2
print("Input1:", repr(input_data1))
print("Input2:", repr(input_data2))
print("Output:", repr(output_data))

get_quantization_params_from_data(np.concatenate([input_data1,input_data2]))
get_quantization_params_from_data(output_data)
```
